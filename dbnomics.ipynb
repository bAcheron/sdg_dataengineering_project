```python
# Main Economic Indicators Publication [MEI] data
targets = ['CP010000','CP020000','CP030000','CP040000','CP040100',
           'CP040200','CP040300','CP040400','CP040500','CP050000',]
data = {}
for target in targets:
    data[target] = fetch_series(f'OECD/MEI/USA.{target}.CTGY.M')
    data[target] = data[target][['original_period','period','original_value']]
```

```python
from plotly import tools
import plotly.graph_objs as go

#Add two Scatter traces to a figure
fig = go.Figure()

for target in targets:
    fig.add_trace(go.Scatter(x=data[target]['original_period'], 
                             y=data[target]['original_value'],
                             name=target))


fig.update_layout(legend_title_text = "Consumer Price Index Contribution")
fig.update_xaxes(title_text="Time")
fig.update_yaxes(title_text="MEI")
fig.show()
```
